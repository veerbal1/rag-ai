```python
import textwrap

text = "My Name is Veerbal Singh. I am a Software Engineer.\na) Chandigarh\nb)California\nc)LA"
img = Image.open('data/1.png')
draw = ImageDraw.Draw(img)
font = ImageFont.truetype("data/garam-light.ttf", 50)

wrapped_text = textwrap.fill(text, width=20) # Set the width as per your requirements.
left, top, right, bottom = draw.textbbox((0, 0), wrapped_text, font=font)
text_width = right - left
text_height = bottom - top

image_width, image_height = img.size

x = (image_width - text_width) // 2
y = (image_height - text_height) // 2
print(text_width, text_height)

new_image = draw.multiline_text((x, y), wrapped_text, fill=(255, 255, 255), font=font, align="center")

plt.imshow(img)
plt.show()
```
